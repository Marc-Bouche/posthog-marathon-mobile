# Additional features

```python
# Sort the dataframe by timestamp before calculating time differences
df = df.sort_values(by='timestamp')

# Now calculate the time differences
df['time_dff'] = df['timestamp'].diff()


# Create 'interaction_id' by detecting when a new interaction occurs based on time and user/app changes
df['interaction_id'] = (((df.timestamp - df.timestamp.shift(1) > pd.Timedelta(1, 'm'))  # Check if the time gap between events is greater than 1 minute
                         & (df.event_type == 'Opened'))  # Ensure the event is 'Opened'
                         | ~(df.app_name == df.app_name.shift(1))  # Detect when the app changes
                         | ~(df.user_id == df.user_id.shift(1))).cumsum()  # Detect when the user changes, then accumulate counts

# Create 'session_id' by detecting when a session ends based on time gap and user changes
df['session_id'] = (((df.timestamp - df.timestamp.shift(1) > pd.Timedelta(5, 'm'))  # Check if the time gap between events is greater than 5 minutes
                     & (df.event_type == 'Opened'))  # Ensure the event is 'Opened'
                     | ~(df.user_id == df.user_id.shift(1))).cumsum()  # Detect when the user changes, then accumulate counts


# Replace NaT with a timedelta of 0 days
df['time_dff'].fillna(pd.Timedelta(0), inplace=True)

# Make sure df_start and df_end are copies, not views
df_start = df.drop_duplicates(subset=['interaction_id'], keep='first').copy()
df_end = df.drop_duplicates(subset=['interaction_id'], keep='last').copy()

# Set index
df_start.set_index('interaction_id', inplace=True)
df_end.set_index('interaction_id', inplace=True)

# Assign timestamps safely using .loc
df_start.loc[:, 'open_time'] = df_start['timestamp']
df_end.loc[:, 'close_time'] = df_end['timestamp']  # Ensure close_time is set for df_end

# Merge start and end timeframes into the main df (using 'interaction_id')
df = df.merge(df_start[['open_time']], on='interaction_id', how='left')
df = df.merge(df_end[['close_time']], on='interaction_id', how='left')

# Result
df
```

# Columns definitions

- **user_id:** A unique identifier for each user.</br>
- **session_id:** A unique identifier for each session. A session starts when:</br>
    -  the user opens an app, and ends when the user closes it</br>
    -  when more than 5 minutes elapsed between two timstamp</br>
    -  when there is a change in user </br>
- **timestamp:** The exact time the event occurred, in datetime format.</br>
- **app_name:** The name of the application the user is interacting with.</br>
- **event_type:** Describes the type of event ("Opened", "Closed", "User Interaction", "Broken").</br>
- **app_category:** The category to which the app belongs (e.g., "Games", "Productivity").</br>
- **time_dff:** The time difference between the current event and the previous event in the session (calculated as timestamp.diff()).</br>
- **interaction_id:** interaction_id represents a unique interaction within a session, capturing specific user actions (like opening or closing an app). It differentiates between multiple interactions within the same session. On the other hand, session_id represents the entire session, which may contain multiple interactions (e.g., app opens and closes) for a single user during a defined period.</br>
- **open_time:** The timestamp when the app was opened during the interaction.</br>
- **close_time:** The timestamp when the app was closed during the interaction.</br>
