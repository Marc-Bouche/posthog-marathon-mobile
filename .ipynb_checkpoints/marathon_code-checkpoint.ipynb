## Format existing dataframe with timestamp as per Posthog csv requirements

```python
import pandas as pd
from datetime import datetime, timedelta

# Load the dataset
df = pd.read_csv('df_all_featured.csv')

# Define race days for each year
race_days = {
    2015: '2015-04-20',
    2016: '2016-04-18',
    2017: '2017-04-17'
}

# Function to create timestamp from race start time and the checkpoint time
def calculate_timestamp(row, checkpoint_time_str, start_time):
    try:
        # Get the race day based on the year
        race_day = race_days.get(row['year'])
        if not race_day:
            return None  # If the race day is not found for the year

        # Parse the race day and checkpoint time as datetime objects
        race_day_datetime = datetime.strptime(race_day, '%Y-%m-%d')
        checkpoint_time = datetime.strptime(checkpoint_time_str, "%H:%M:%S")
        
        # Calculate the time difference between the checkpoint and the start time
        delta = timedelta(hours=checkpoint_time.hour, minutes=checkpoint_time.minute, seconds=checkpoint_time.second)
        timestamp = race_day_datetime + delta
        return timestamp.strftime('%Y-%m-%dT%H:%M:%SZ')
    except Exception as e:
        print(f"Error processing timestamp for {checkpoint_time_str}: {e}")
        return None

# Restructure the DataFrame for event-based tracking
event_data = []

# Loop through each row in the DataFrame
for _, row in df.iterrows():
    runner_id = row['id']
    checkpoints = ['5K', '10K', '15K', '20K', 'half', '25K', '30K', '35K', '40K']
    
    # Process checkpoint times
    for checkpoint in checkpoints:
        checkpoint_time = row[checkpoint]
        
        if pd.notna(checkpoint_time):  # Skip if the checkpoint time is missing
            timestamp = calculate_timestamp(row, checkpoint_time, start_time=datetime.strptime("10:00:00", "%H:%M:%S"))
            event_data.append({
                'event': 'runner_checkpoint',
                'properties': {
                    'runner_id': runner_id,
                    'checkpoint': checkpoint,
                    'time': checkpoint_time,
                    'timestamp': timestamp
                }
            })
    
    # Add the final time event for each runner
    final_time = row['time']
    if pd.notna(final_time):  # Skip if the final time is missing
        event_data.append({
            'event': 'runner_finish',
            'properties': {
                'runner_id': runner_id,
                'checkpoint': 'finish',  # Set checkpoint to 'finish'
                'time': final_time,
                'timestamp': calculate_timestamp(row, final_time, start_time=datetime.strptime("10:00:00", "%H:%M:%S"))
            }
        })

# Convert the event data into a DataFrame
event_df = pd.DataFrame(event_data)

# Expand the 'properties' column into individual columns
event_df_expanded = pd.json_normalize(event_df['properties'])

# Combine with the 'event' column to have a complete structure
final_event_df = pd.concat([event_df[['event']], event_df_expanded], axis=1)

# Display the final DataFrame
final_event_df.head(30)
```

```python
# Fill the missing timestamp with the current time or a placeholder
final_event_df['timestamp'] = final_event_df['timestamp'].fillna(pd.to_datetime('now').strftime('%Y-%m-%dT%H:%M:%SZ'))

# Verify that the missing value is handled
final_event_df.isnull().sum()
```
